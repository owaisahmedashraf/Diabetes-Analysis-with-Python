```python
#Question8:

new_list=[]
def norm_bp(column_name):
    a = df[column_name].to_numpy()
    b = np.mean(a)
    c = np.std(a)
    for i in a:
        norm = ((i - b)/c)
        new_list.append(norm)
    return(new_list)


```

```python
#Question 4:

def weight_class(bmi_value):
        if bmi_value < 18.5:
            return("Underweight")
        elif 18.5 <= bmi_value <= 24.9:
            return("Normal")
        elif 25 <= bmi_value <= 29.9:
            return("Overweight")
        else:
            return("Obese")
        
```
